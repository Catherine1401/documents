```python
# Load data
# (1) Upload file lên Google colab
# Tỉ lệ nam / nữ
male_count = 0
female_count = 0

with open("Social_Network_Ads.csv", "r") as f:
  data = f.readlines()
  print(type(data))
  for line in data[1:]:
    line = line.strip()
    # loại bỏ khoảng trắng và ký tự đặc biệt
    # ở đầu hàng và cuối hàng
    line = line.split(',')
    # tách line thành các sub-string
    # lấy dấu "," làm dấu hiệu để tách
    if line[1] == "Male":
      male_count += 1
    else:
      female_count += 1
print(male_count, female_count)

```

```python
import matplotlib.pyplot as plt # vẽ đồ thị
import numpy as np # tính toán số học cơ bản
import pandas as pd # phân tích dữ liệu

x = ["Male", "Female"]
y = [male_count, female_count]
plt.bar(x, y, width=0.3)
plt.xlabel("Gender", fontsize=15)
plt.ylabel("Count", fontsize=15)
```

```python
# Tính tuổi trung bình và thu nhập trung bình

s = 0
count = 0

with open("Social_Network_Ads.csv", "r") as f:
  data = f.readlines()

for line in data[1:]:
  line = line.strip()
  line = line.split(',')
  if line[1] == "Female":
    s += float(line[3]) # string -> intege
    count += 1

print(f"Thu nhập trung bình {s/count} USD / Year")
```
